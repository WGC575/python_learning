```python
#We use two-sided fisher exact test to analyze whether there is significant difference between types (29 types) of commits (by purpose) and whether they witness a increase of software metrics.

#PMD
import scipy.stats as ss
import pandas as pd
import numpy

neutral = pd.read_csv('./neutral_metric.csv',)
metric = pd.read_csv('./sq_2.csv')

for j in metric.columns:
    for i in neutral.columns:
        print(ss.fisher_exact([[neutral[i][0],neutral[i][1]],[metric[j][0], metric[j][1]]],"two-sided")[1])
    print("\n")
```
